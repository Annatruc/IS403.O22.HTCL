```python
import pandas as pd
df=pd.read_csv('insurance_survey.csv')
df
```

```python
df_gr = df.groupby('Education')['Satisfaction* '].apply(list)
df_gr
```

```python
from scipy.stats import levene

stat, p = levene(*df_gr, center='mean')
stat, p
```

```python
from scipy.stats import f_oneway

fvalue, pvalue = f_oneway(*df_gr)
fvalue, pvalue
```
