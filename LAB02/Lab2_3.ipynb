```python
import pandas as pd
from scipy.stats import levene
from statsmodels.stats.multicomp import pairwise_tukeyhsd
df=pd.read_csv('Exercise_4_Final.csv')
df
```

```python
# repaired_df = pd.melt(df, var_name='Khoi', value_name='Diem')
repaired_df = df.groupby('Khoi')['Diem'].apply(list)
repaired_df
```

```python
st, pval = levene(*repaired_df, center="mean")
st, pval
```
