```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
data = pd.read_csv("DATA_2B.csv")
data
```

```python
x = np.array(data[['log(seating)', 'log(year)']]).reshape((-1, 2))
y = np.array(data['price. Price']).reshape((-1, 1))
x, y
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
reg.intercept_
```

```python
reg.coef_
```

```python
reg.score(x,y)
```

```python

y = data['price. Price']

# For independent variables (x), select the desired columns
x = data[['log(seating)', 'log(year)']]

# Add a constant term to the independent variables
x = sm.add_constant(x)

reg = sm.OLS(y,x).fit()
reg.summary()
```
